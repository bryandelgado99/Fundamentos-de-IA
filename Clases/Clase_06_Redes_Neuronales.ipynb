# Ejercicio 1 - TransformaciÃ³n de temperatura

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
import pandas as pd

# Leemos el archivo
datos = pd.read_excel("/content/temperatura.xlsx")
print(datos.head())

# Separamos los datos de entrada y salida
X=datos["celsius"]
Y=datos["fahrenheit"]

# Dividimos datos en aprendizaje y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=25)

#Creamos la red neuronal de tipo percetrÃ³n
capa_entrada = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa_entrada])

#Compilar el modelo
modelo.compile(optimizer='adam', loss='mean_squared_error')

#Entrenamiento del modelo
historial = modelo.fit(X_train, Y_train, epochs=100, batch_size=1)
print(historial)
```

```python
#ImpresiÃ³n de historial
import matplotlib.pyplot as plt
plt.plot(historial.history["loss"])

#FunciÃ³n de pÃ©rdida
perdida = modelo.evaluate(X_test, Y_test)
print(perdida)

print(modelo.predict([25]))
```
