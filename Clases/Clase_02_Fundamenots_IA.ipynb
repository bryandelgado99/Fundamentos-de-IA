# **Ejercicio 1: WSHouse**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

#Leer archivo
datos=pd.read_excel("/content/WSHouse.xlsx")
#Muestra los datos
datos.head()
#Describe las características del datasheet
datos.describe()
```

```python
#Separar datos de entrada X con características y Y con etiquetas
x = datos[['Tamanio']]
y = datos[['Precio_venta']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 25)

#Creamos el modelo
modelo = LinearRegression()

#Entrenamiento del modelo
modelo.fit(x_train,y_train)

#Probar el modelo y obtener datos
y_obtenido = modelo.predict(x_test)

#Métricas de evaluación
error = mean_squared_error(y_test, y_obtenido)
print(f"El error es: {error}")

#Coeficiente de determinación
r2 = r2_score(y_test, y_obtenido)
print(f"El coeficiente de determinación es: {r2}")
```

```python
#Nueva predicción
precio = modelo.predict([[2]])
print(f"El precio es: {precio}")

#Imprimimos los valores
print(f"El coeficiente B1 es: {modelo.coef_}")
print(f"La Intercepción es: {modelo.intercept_}")
```

```python
#Gráfica
import matplotlib.pyplot as plt

plt.scatter(x['Tamanio'], y)
plt.title("Regresión Lineal")
plt.xlabel("Tamaño")
plt.ylabel("Precio Adecuado")
plt.show()
```

# **Ejercicio 2: Whine Red Queality**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

#Leer archivo
datos=pd.read_csv("/content/winequality_red.csv")
#Muestra los datos
datos.head()
#Describe las características del datasheet
datos.describe()
```

```python
#Separar datos de entrada X con características y Y con etiquetas
x = datos.iloc[:,:-1] #Selecciona todas las columnas menos la última
y = datos.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 25)

#Creamos el modelo
modelo = LinearRegression()

#Entrenamiento del modelo
modelo.fit(x_train,y_train)

#Probar el modelo y obtener datos
y_obtenido = modelo.predict(x_test)

#Métricas de evaluación
error = mean_squared_error(y_test, y_obtenido)
print(f"El error es: {error}")

#Coeficiente de determinación
r2 = r2_score(y_test, y_obtenido)
print(f"El coeficiente de determinación es: {r2}")

#Nueva predicción
precio = modelo.predict([[2, 5.0, 9.0, 6, 0.1, 0.1, 7, 0.2, 9, 0.01, 5]])
print(f"El precio es: {precio}")

#Imprimimos los valores
print(f"El coeficiente B1 es: {modelo.coef_}")
print(f"La Intercepción es: {modelo.intercept_}")
```

```python
#Gráfica de dispersión
import matplotlib.pyplot as plt

plt.scatter(x['density'], y)
plt.title("Regresión Lineal")
plt.xlabel("Densidad")
plt.ylabel("Calidad")
plt.show()

#Gráfica de valores obtenidos vs valores reales
#Creamos la lista desde 1 hasta la longitud de y_test
lista = [i for i in range(1, len(y_test)+1, 1)]
plt.plot(lista, y_test, color='r')
plt.plot(lista, y_obtenido, color='g')
plt.show()
```
